```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# descarga de datos
url_datos="https://raw.githubusercontent.com/NechuBM/dataset/main/precios_casa.txt"
!wget {url_datos} -q -O precios_casa.txt

df = pd.read_csv('precios_casa.txt', delimiter=';')
```

```python
df.head()
```

```python
df.barrio.unique()
```

```python
df_multiple = pd.get_dummies(df,'barrio',drop_first=True)
```

```python
df_multiple['interseccion']=1
x=df_multiple[['interseccion','tamano','antiguedad','barrio_N']]
y=df_multiple.precio
```

```python
x = x.astype(float)
y = y.astype(float)
```

```python
df_multiple.head()
```

```python
ml = sm.OLS(y, x)
resultado = ml.fit()
print(resultado.summary())
```
